```python
import numpy as np
import pandas as pd
```

```python
Corpus = pd.read_csv("/content/tagged02-07-2022-122843.csv",encoding='utf8')
```

```python
Corpus['Tag'] = [int(str(entry).replace('2','0')) for entry in Corpus['Tag']]
Corpus['Tag'] = [int(str(entry).replace('4','0')) for entry in Corpus['Tag']]
Corpus['Tag'] = [int(str(entry).replace('3','1')) for entry in Corpus['Tag']]
```

```python
Corpus['dcNoteFinal'] = [entry.lower().replace('b"','') for entry in Corpus['dcNoteFinal']]
```

```python
import nltk
nltk.download('punkt')
```

```python
import re
Corpus['dcNoteFinal'] = [(re.sub(r'[^\w\s]','',entry)) for entry in Corpus['dcNoteFinal']]
```

```python
Corpus['dcNoteFinal'] = [entry.replace('\t',' ') for entry in Corpus['dcNoteFinal']]
Corpus['dcNoteFinal'] = [entry.replace('    ',' ') for entry in Corpus['dcNoteFinal']]
Corpus['dcNoteFinal'] = [entry.replace('   ',' ') for entry in Corpus['dcNoteFinal']]
Corpus['dcNoteFinal'] = [entry.replace('  ',' ') for entry in Corpus['dcNoteFinal']]
```

```python
Corpus['dcNoteFinal']
```

```python
from sklearn import model_selection
Train_X, Test_X, Train_Y, Test_Y = model_selection.train_test_split(Corpus['dcNoteFinal'],Corpus['Tag'],test_size=0.3)
```

```python
train_x = Train_X.to_frame()
train_y = Train_Y.to_frame()
test_x = Test_X.to_frame()
test_y = Test_Y.to_frame()
```

```python
train_x
```

```python
train_y
```

```python
train_bio = pd.DataFrame.join(train_x, train_y,how='left')
```

```python
test_bio = pd.DataFrame.join(test_x, test_y,how='left')
```

```python
train_bio
```

```python
import pandas as pd

with open("/content/test.txt",'a') as f:
  f.write('index' + '\t' + 'sentence' + '\t' + 'label' + '\n')
  for i in range(len(test_bio)):
    row = test_bio.iloc[i].values.tolist()
    str_i = str(i)
    f.write(str_i + '\t' + row[0] + '\t' + str(row[1]) + '\n')
```
